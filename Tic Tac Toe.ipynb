```python
def turn(sym):
    k = int(input('Where do you wish to place your marker?\n At position\t'))
    return k
    
```

```python
def print_board(pos, sym, str1):
    
    str1[pos-1] = sym.upper()
    answer_string = (f'   |   |   \n {str1[0]} | {str1[1]} | {str1[2]} \n___|___|___\n   |   |   \n {str1[3]} | {str1[4]} | {str1[5]} \n___|___|___\n   |   |   \n {str1[6]} | {str1[7]} | {str1[8]} \n   |   |   ')
    print('\n'*100)
    print(answer_string)
    return str1
```

```python
# Select Player Number 
p = int(input('Do you want to be Player 1 or 2\n Player :\t'))

if p == 1:
    s = 'X'
else:
    s = 'O'

#Choose position of marker
if s.upper() == 'X' or s.upper() == 'O':
    position = turn(s)
else:
    print('\nEnter valid symbol and try again\n')

#Print the board    
str1 = [' ',' ',' ',' ',' ',' ',' ',' ',' ']

ans_str = print_board(position, s, str1)

for x in range(8):
    if p ==1:
        if x % 2 == 0:
            s = 'O'
        else:
            s = 'X'
    else:
        if x % 2 == 0:
            s = 'X'
        else:
            s = 'O'
    
    position = turn(s)
    ans_str = print_board(position, s, ans_str)
    lix = ['X','X','X']
    lio = ['O','O','O']
    if [ans_str[0], ans_str[1], ans_str[2]] == lix or [ans_str[3], ans_str[4], ans_str[5]] == lix or [ans_str[6], ans_str[7], ans_str[8]] == lix or [ans_str[1], ans_str[4], ans_str[7]] == lix or [ans_str[0], ans_str[3], ans_str[6]] == lix or [ans_str[2], ans_str[5], ans_str[8]] == lix or [ans_str[0], ans_str[4], ans_str[8]] == lix or [ans_str[2], ans_str[4], ans_str[6]] == lix:
        print('\nPlayer 1 is the winner!')
        break
    elif [ans_str[0], ans_str[1], ans_str[2]] == lio or [ans_str[3], ans_str[4], ans_str[5]] == lio or [ans_str[6], ans_str[7], ans_str[8]] == lio or [ans_str[1], ans_str[4], ans_str[7]] == lio or [ans_str[0], ans_str[3], ans_str[6]] == lio or [ans_str[2], ans_str[5], ans_str[8]] == lio or [ans_str[0], ans_str[4], ans_str[8]] == lio or [ans_str[2], ans_str[4], ans_str[6]] == lio:
        print('\nPlayer 2 is the winner!')
        break
    else:
        continue
else:
    print('\nIt\'s a draw!')
    
    
    
```
